```python
import psycopg2
import config

def view_taste_profile():
    db_config = {
        "dbname": config.DB_NAME,
        "user": config.DB_USER,
        "password": config.DB_PASS,
        "host": config.DB_HOST,
        "port": config.DB_PORT
    }
    conn = psycopg2.connect(**db_config)
    cur = conn.cursor()
    cur.execute("SELECT * FROM taste_profile;")
    rows = cur.fetchall()
    for row in rows:
        print(row)
    cur.close()
    conn.close()

if __name__ == "__main__":
    view_taste_profile()
```

```python
#!/usr/bin/env python
from db_functions import clear_table, get_all_table_names

def clear_all_tables():
    """
    Clears all tables in the database one by one and reports the results.
    """
    # Known tables from the code
    tables_to_clear = [
        'inventory',
        'taste_profile',
        'meals',
        'new_meal_ideas',
        'meal_ideas_in_stock',
        'saved_meals_in_stock'
    ]
    
    # Get all tables from database to ensure we don't miss any
    all_tables = get_all_table_names()
    
    # Combine and remove duplicates
    all_tables = list(set(tables_to_clear + all_tables))
    
    results = []
    for table in all_tables:
        success = clear_table(table)
        status = "â" if success else "â"
        results.append(f"{status} {table}")
    
    # Print results
    print("\nTable clearing results:")
    print("----------------------")
    for result in results:
        print(result)

if __name__ == "__main__":
    clear_all_tables()
```

```python
from db_functions import get_daily_notes_range
from meal_plan_in_stock_checker import MealPlanInStockChecker
from datetime import datetime

# Initialize checker and get current date
checker = MealPlanInStockChecker()
current_date = datetime.now().strftime('%Y-%m-%d')

# Get notes for today
notes_for_day = get_daily_notes_range(current_date, current_date)
missing_ingredients = set()

# Process today's note if it exists
if notes_for_day and notes_for_day[0]:
    _, _, note_content = notes_for_day[0]
    missing_ingredients.update(checker.get_missing_ingredients(note_content))

# Display results
print("\nShopping List Generated Successfully!")
print("\nMissing Ingredients:")
if missing_ingredients:
    for ingredient in sorted(missing_ingredients):
        print(f"- {ingredient}")
else:
    print("No missing ingredients found.")
```
