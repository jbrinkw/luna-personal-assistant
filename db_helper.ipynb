```python
#!/usr/bin/env python
# db_helper_fixed.py

"""
This script displays the contents of the database tables.
It has been updated to support the new taste_profile table structure
that only has a profile column (no id column).
"""

from debug.reset_db import ResetDB
from datetime import datetime, timedelta

def display_db_contents():
    resetdb = ResetDB()
    db = resetdb.db
    tables = resetdb.tables
    
    # Display each table and its records
    for table_name, table_obj in tables.items():
        print(f"\n*** {table_name.upper()} ***")
        records = table_obj.read()
        
        if not records:
            print("  No records found")
            continue
            
        print(f"  Total records: {len(records)}")
        
        # Display first few records for each table
        if table_name == "daily_planner":
            # Show 7 days starting from 3/25/25
            start_date = datetime(2025, 3, 25).date()
            for i in range(7):
                current_date = start_date + timedelta(days=i)
                record = None
                for r in records:
                    if r[0] == current_date:
                        record = r
                        break
                        
                print(f"\n  Date: {current_date}")
                if record:
                    print(f"  Notes: {record[1]}")
                    print(f"  Meal IDs: {record[2]}")
                else:
                    print("  No plans for this day")
                    
        else:
            for i, record in enumerate(records[:5]):  # Show first 5 records
                print(f"\n  Record {i+1}:")
                
                if table_name == "inventory":
                    print(f"  ID: {record[0]}")
                    print(f"  Name: {record[1]}")
                    print(f"  Quantity: {record[2]}")
                    print(f"  Expiration: {record[3]}")
                elif table_name == "taste_profile":
                    # Updated to handle the new structure with only the profile column
                    print(f"  Profile: {record[0]}")
                elif table_name in ["saved_meals", "new_meal_ideas"]:
                    print(f"  ID: {record[0]}")
                    print(f"  Name: {record[1]}")
                    print(f"  Prep Time: {record[2]} minutes")
                    print(f"  Ingredients: {record[3]}")
                    print(f"  Recipe: {record[4]}")
                elif table_name == "shopping_list":
                    print(f"  ID: {record[0]}")
                    print(f"  Amount: {record[1]}")
                elif table_name == "ingredients_foods":
                    print(f"  ID: {record[0]}")
                    print(f"  Name: {record[1]}")
                    print(f"  Min Amount: {record[2]}")
                    print(f"  Walmart Link: {record[3]}")
                else:
                    # Generic display for other tables
                    for j, value in enumerate(record):
                        print(f"  Column {j+1}: {value}")
                        
            if len(records) > 5:
                print(f"\n  ... and {len(records) - 5} more records")

if __name__ == "__main__":
    # Display the database contents
    display_db_contents()
```
